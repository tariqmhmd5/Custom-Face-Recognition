```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
```

```python
model = load_model('facefeatures_new_model.h5')       #import model weights
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')  #import face classifier
```

```python
def face_extractor(img):
    # Function detects faces and returns the cropped face
    # If no face detected, it returns the input image
    
    #gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(img, 1.3, 5)
    
    if faces is ():
        return None
    # Crop all faces found
    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,255),2)
        cropped_face = img[y:y+h, x:x+w]

    return cropped_face
```

```python
# Testing model on a single image
im = cv2.imread('00002524-12.jpg')
im = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
plt.imshow(im)
pred = model.predict(im.reshape(1,96,96,3))
for i in range(25):
    if pred[0][i]>0.5:
        print(i)
# as you can output is pred[0][9] which is belong to the class 0012 folder we can cross check
```
